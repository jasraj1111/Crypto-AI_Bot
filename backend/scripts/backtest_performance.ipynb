```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load backtest results
df_backtest = pd.read_csv("data/backtest_results.csv")

# Ensure timestamp is in datetime format
df_backtest["timestamp"] = pd.to_datetime(df_backtest["timestamp"])

# Calculate cumulative returns
df_backtest["cumulative_return"] = (1 + df_backtest["profit_loss"]).cumprod()

# Plot cumulative returns
plt.figure(figsize=(12, 6))
plt.plot(df_backtest["timestamp"], df_backtest["cumulative_return"], label="Cumulative Return", color="blue")
plt.axhline(y=1, color="gray", linestyle="--")  # Baseline
plt.xlabel("Timestamp")
plt.ylabel("Cumulative Return")
plt.title("Cumulative Trading Returns Over Time")
plt.legend()
plt.grid()
plt.show()
```
